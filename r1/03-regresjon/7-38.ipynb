```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data3 = np.array([4.43, 3.96, 4.03, 3.68, 4.31])
data6 = np.array([2.69, 2.88, 2.78, 2.86, 3.10])
data9 = np.array([2.38, 2.28, 2.44, 2.21, 2.51])
data12 = np.array([1.75, 1.91, 1.99, 2.11, 2.09])
```

```python
sum(data3) / len(data3)
```

```python
avg3 = np.average(data3)
avg6 = np.average(data6)
avg9 = np.average(data9)
avg12 = np.average(data12)

xs = np.array([3, 6, 9, 12])
ys = np.array([avg3, avg6, avg9, avg12])
```

```python
def model_exp(t, a, b):
    return a * b ** t
```

```python
coeffs, _ = curve_fit(model_exp, xs, ys)
```

```python
xs_model = np.linspace(0, 25, 1000)
ys_model = model_exp(xs_model, *coeffs)

plt.scatter(xs, ys)
plt.plot(xs_model, ys_model)
plt.show()
```
