# Numerisk derivasjon

Husk at den deriverte til en funksjon $f$ i et punkt $a$ er gitt ved $\displaystyle f' (a) = \lim_{h \to 0} \frac{f(a-h) - f(a)}{h}$

```python
f = lambda x: 3 * x ** 2 - 2 * x + 1
```

## Enkel approksimasjon

Bruke definisjonen direkte

```python
def derivert(f, x, dx=1E-4):
    return (f(x + dx) - f(x)) / dx﻿
```

```python
derivert(f, 1)
```

## Newtons kvotient

Forskjell: Ser på området fra $a-h$ til $a+h$.

Endring i $x$-retning er nå $2h$

```python
def derivert_newton(f, x, dx=1E-4):
    return (f(x+h) - f(x-h)) / (2 * h)
```

```python
derivert_newton(f, 1)
```

## Oppgave

$$
f(x) = \sqrt{\ln{x}}
$$

```python
import numpy as np
f = lambda x: np.sqrt(np.log(x))

dy1 = derivert(f, 3)
dy2 = derivert_newton(f, 3)

dy1, dy2
```

## Plotting av funksjoner

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 3 * x ** 2 - 2 * x + 1

x = np.linspace(-1, 4, 100)
y = f(x)
dy = derivert_newton(f, x)

plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y, label="f(x)", color="limegreen", linewidth=5)
plt.plot(x, dy, label="f'(x)", color="salmon", linewidth=5)
plt.legend()
plt.grid()
plt.axhline(color="gray")
plt.axvline(color="gray")
plt.show()
```
