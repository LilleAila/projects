```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def model(t, a, b, C):
    return C / (1 + a * np.exp(-b * t))
```

```python
x = [0, 8, 20, 28, 39, 54, 69, 84, 100]
y = [1, 21, 1118, 2580, 5107, 6890, 7787, 8823, 8433]

plt.scatter(x, y)
```

```python
coeffs, _ = curve_fit(model, x, y, p0=[0, 0, 8433])

coeffs
```

```python
xs = np.linspace(0, 100, 1000)
ys = model(xs, *coeffs)

plt.plot(xs, ys, color="red", label="Modell")
plt.scatter(x, y, color="navy", label="Datapunkter")
plt.grid()
plt.legend()
plt.show()
```
