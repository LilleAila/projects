# Analyse av lesing fra koreader

Gjort ved hjelp av python, uten bruk av pandas.

```python
import matplotlib.pyplot as plt
import csv
```

```python
with open("books.csv") as file:
    reader = csv.DictReader(file)
    books = list(reader)
```

```python
with open("pages.csv") as file:
    reader = csv.DictReader(file)
    pages = list(reader)
```

```python
data = sorted([
    (b["title"], float(b["total_read_time"]) / float(b["total_read_pages"]))
    for b in books
], key=lambda b: b[1])

fig, ax = plt.subplots()
ax.barh(*zip(*data))
ax.set_title("Books by time spent per page")
ax.set_xlabel("Seconds per page")
ax.set_ylabel("")
```

```python
data = sorted([
    (b["title"], float(b["total_read_time"]) / (60*60))
    for b in books
], key=lambda b: b[1])

fig, ax = plt.subplots()
ax.barh(*zip(*data))
ax.set_title("Total time spent reading by book")
ax.set_xlabel("Time spent reading in hours")
ax.set_ylabel("")
```
