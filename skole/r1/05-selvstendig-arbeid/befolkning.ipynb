```python
import csv
from matplotlib import pyplot as plt
```

```python
with open("befolkning-aldersgrupper-1986-2025.csv", encoding="latin1") as file:
    reader = csv.DictReader(file, delimiter=";")
    df = [
        {
            "kjønn": r["kjønn"],
            # NOTE: 105 inkluderer alle eldre
            "alder": int(r["alder"]),
            "år": int(r["år"]),
            "personer": int(r["Personer"])
        }
        for r in reader
    ]
        
df[:10]
```

```python
for a in range(106):
    values = [
        (r["år"], r["personer"])
        for r in df
        if r["kjønn"] == "Kvinner" and r["alder"] == a
    ]
    plt.plot(*zip(*values), label=f"{a} år gamle kvinner")
plt.legend()
```
