```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def model_exp(t, A0, b):
    return A0 * b ** t

def model_e(t, A0, k):
    return A0 * np.exp(k * t)
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 7]
y = [200, 147, 112, 80, 55, 42, 32, 25]

plt.scatter(x, y)
```

```python
coeffs_exp, _ = curve_fit(model_exp, x, y)
coeffs_e, _ = curve_fit(model_e, x, y)

coeffs_exp, coeffs_e
```

```python
xs = np.linspace(0, 7, 1000)
ys_exp = model_exp(xs, *coeffs_exp)
ys_e = model_e(xs, *coeffs_e)

plt.plot(xs, ys_exp, color="red", label="Modell 1")
plt.plot(xs, ys_e, color="green", label="Modell 2")
plt.scatter(x, y, color="navy", label="Datapunkter")
plt.grid()
plt.legend()
plt.show()
```
